```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('ripon.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.show()
```

```python
# Display the images
plt.hist(img.ravel(),256,range = [0, 256]);
plt.title('Original Image')
plt.show()
```

```python
# Equalize histogram
img_eq = cv2.equalizeHist(img)
```

```python
# Display the images.
plt.hist(img_eq.ravel(), 256, range = [0, 256])
plt.title('Equalized Histogram')
```

```python
# Display the images.
plt.imshow(img_eq, cmap='gray')
plt.title('Original Image')
plt.show()
```

# Histogram Equalization for Color Images

* WORKFLOW 
- Tranform the image to HSV colorspace.
- Perform histogram equalization only on the V channel.
- Transform the image back to RGB colorspace.

```python
# Read the color image.
img = cv2.imread('ripon.jpg', cv2.IMREAD_COLOR)
```

```python
# Convert to HSV.
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
# Perform histogram equalization only on the V channel, for value intensity.
img_hsv[:,:,2] = cv2.equalizeHist(img_hsv[:, :, 2])
```

```python
# Convert back to BGR format.
img_eq = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2BGR)
```

```python
plt.imshow(img_eq[:,:,::-1]) 
plt.title('Equalized Image')
plt.show()
```

```python
plt.hist(img_eq.ravel(),256,range = [0, 256])
plt.title('Histogram Equalized')
plt.show()
```

```python
# Display the images.
#plt.figure(figsize = (20,10))
plt.subplot(221)
plt.imshow(img[:, :, ::-1])
plt.title('Original Color Image')
plt.subplot(222)
plt.imshow(img_eq[:, :, ::-1])
plt.title('Equalized Image')
plt.subplot(223)
plt.hist(img.ravel(),256,range = [0, 256])
plt.title('Original Image')
plt.subplot(224)
plt.hist(img_eq.ravel(),256,range = [0, 256])
plt.title('Histogram Equalized');plt.show()
```
